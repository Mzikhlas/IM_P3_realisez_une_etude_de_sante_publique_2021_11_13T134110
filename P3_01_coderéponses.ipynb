```python
#Question2 Liens entre production,importations, exportations, disponibilité intérieure et variations de stock
#Il semblerait que la disponibilité intérieure se calcule de la façon suivante :
#Dispo_interieur == (Production + Importation - Exportation  + Variation_stock) == (Nourriture + Nourriture_animaux + Semences + Pertes + Traitement + Autres_utilisations_non_alimentaire)
```

```python
frvegetaux['element'].unique()
```

```python
Production = frvegetaux.loc[frvegetaux['element'] == 'Production'].sum()['quantite']*1000000
Importation = frvegetaux.loc[frvegetaux['element'] == 'Importations - Quantité'].sum()['quantite']*1000000
Exportation = frvegetaux.loc[frvegetaux['element'] == 'Exportations - Quantité'].sum()['quantite']*1000000
Variation_stock = frvegetaux.loc[frvegetaux['element'] == 'Variation de stock'].sum()['quantite']*1000000
Dispo_interieur = Production + Importation - Exportation  + Variation_stock
Dispo_interieur
```

```python
Nourriture = frvegetaux.loc[frvegetaux['element'] == 'Nourriture'].sum()['quantite']*1000000
Nourriture_animaux = frvegetaux.loc[frvegetaux['element'] == 'Aliments pour animaux'].sum()['quantite']*1000000
Semences = frvegetaux.loc[frvegetaux['element'] == 'Semences'].sum()['quantite']*1000000
Pertes = frvegetaux.loc[frvegetaux['element'] == 'Pertes'].sum()['quantite']*1000000
Traitement = frvegetaux.loc[frvegetaux['element'] == 'Traitement'].sum()['quantite']*1000000
Autres_utilisations_non_alimentaire = frvegetaux.loc[frvegetaux['element'] == 'Autres utilisations (non alimentaire)'].sum()['quantite']*1000000
Dispo_interieur = Nourriture + Nourriture_animaux + Semences + Pertes + Traitement + Autres_utilisations_non_alimentaire
Dispo_interieur
```

```python
Dispo_interieur == (Production + Importation - Exportation  + Variation_stock) == (Nourriture + Nourriture_animaux + Semences + Pertes + Traitement + Autres_utilisations_non_alimentaire)
```

Disponibilité intérieure est donc la redondance qui se retrouve parmis les 11 quantités données

Question 3 : Calculez (pour chaque pays et chaque produit) la disponibilité alimentaire en kcal puis en kg de protéines.

Vous ferez cela à partir de ces informations :

Population de chaque pays ; Disponibilité alimentaire donnée pour chaque produit et pour chaque pays en kcal/personne/jour. Disponibilité alimentaire en protéines donnée pour chaque produit et pour chaque pays en g/personne/jour.

```python
#calcul(pour chaque pays et chaque produit) la disponibilité alimentaire en kcal puis en kg de protéines.
fr_alim.head()
```

```python
#Déduction des valeurs en kcal

JR_AN= 365

fr_alim["Disponibilité alimentaire Kcal"] = fr_alim["Disponibilité alimentaire (Kcal/personne/jour)"] * (fr_alim["population"] * 1000) * JR_AN


#Déduction des valeurs en kg de protéine

fr_alim["Disponibilité kg de protéine"] = fr_alim["Disponibilité de protéines en quantité (g/personne/jour)"] / 1000 * (fr_alim["population"] * 1000) * JR_AN
```

```python
fr_alim.head()
```

On a calculé grâce aux conversions d'unité la disponibilité alimentaire par pays et par produit en kcal uniquement( il faut comme enlevé le "par personne par jour" en calculant)idem pour les protéines en passant du g au kg.

Question 8 : Combien d'humains pourraient être nourris si toute la disponibilité alimentaire en produits végétaux la nourriture végétale destinée aux animaux et les pertes de produits végétaux étaient utilisés pour de la nourriture ? Donnez les résultats en termes de calories, puis de protéines, et exprimez ensuite ces 2 résultats en pourcentage de la population mondiale.

Création d'une fonction pour répondre plus vite aux questions similaires à la Q7 : en terme de Kcal, Kg de protéine et le résultat en pourcentage de la population mondiale

```python
def calculs_populations (donnee_kcal, donnee_kg_prot):
    disp_ajr_energetique_an = 2500 * 365
    disp_ajr_proteines_an = (45/1000)*365
    
    disp_pop_nourrie_kcal = (donnee_kcal / disp_ajr_energetique_an)
    disp_pop_nourrie_prot_kg = (donnee_kg_prot / disp_ajr_proteines_an)
    
    print_calc_pop = pd.DataFrame({'Nombre d\'habitants nourris' : [disp_pop_nourrie_kcal, disp_pop_nourrie_prot_kg],
                         '% de la population nourris' : [(disp_pop_nourrie_kcal / population_2013)*100, 
                                                (disp_pop_nourrie_prot_kg / population_2013)*100]},
                        index = ['Basé dur la dispo en Kcal', 'Basé dur la dispo en Kg de protéines'])
    return print_calc_pop
    
```

On ajoute 2 colonnes au DataFrame fr_vegetal afin de répondre à la question.



```python
fr_vegetal['Dispo Alimentaire q8 (Kcal)'] = ((fr_vegetal['Disponibilité alimentaire Kcal'])
                                                               + ((fr_vegetal['Aliments pour animaux'] * 1000000) 
                                                                  * fr_vegetal['ratio energie/poids'])
                                                               + ((fr_vegetal['Pertes'] * 1000000) 
                                                                  * fr_vegetal['ratio energie/poids'])
                                                              )

fr_vegetal['Dispo Alimentaire Kg protéine q8'] = (fr_vegetal['Disponibilité kg de protéine']
                                                                      +  (((fr_vegetal['Aliments pour animaux'] 
                                                                         + fr_vegetal['Pertes']) * 1000000)
                                                                       * (fr_vegetal['ratio de protéine en %']/100))
                                                                      
                                                                        )
```

Agrégat sur produit avec un groupby pour avoir des résultats au niveau mondial

```python
dispo_alim_vegetaux_produits = fr_vegetal.groupby(by = 'Produit').sum().reset_index()

fr_vegetal.head()
```

On calcul les sommes au niveau mondial

```python
dispo_alim_mondial_vegetaux_kcal = dispo_alim_vegetaux_produits['Dispo Alimentaire q8 (Kcal)'].sum()
dispo_alim_mondial_vegetaux_kgprot = dispo_alim_vegetaux_produits['Dispo Alimentaire Kg protéine q8'].sum()
```

```python
# Appel de la fonction pour obtenir les résultats

calculs_populations(dispo_alim_mondial_vegetaux_kcal,dispo_alim_mondial_vegetaux_kgprot)
```
